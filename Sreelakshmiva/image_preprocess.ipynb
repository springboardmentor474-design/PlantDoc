```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ===============================
# Image Settings
# ===============================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# ===============================
# Dataset Paths
# ===============================
train_dir = "Combined_Dataset/train"
test_dir = "Combined_Dataset/test"

# ===============================
# Training Data Generator (WITH augmentation)
# ===============================
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,        # Rotate images
    width_shift_range=0.2,    # Horizontal shift
    height_shift_range=0.2,   # Vertical shift
    zoom_range=0.2,           # Zoom
    shear_range=0.2,          # Shear
    horizontal_flip=True,     # Flip images
    fill_mode='nearest',
    validation_split=0.2      # 20% validation split
)

# ===============================
# Validation Data Generator (NO augmentation)
# ===============================
val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# ===============================
# Training Generator
# ===============================
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

# ===============================
# Validation Generator
# ===============================
val_generator = val_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

# ===============================
# Test Generator (NO augmentation)
# ===============================
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# ===============================
# Print Dataset Summary
# ===============================
print("\nTraining Images:", train_generator.samples)
print("Validation Images:", val_generator.samples)
print("Test Images:", test_generator.samples)

print("\nNumber of Classes:", train_generator.num_classes)

print("\nClass Indices:")
print(train_generator.class_indices)

# ===============================
# Check One Batch
# ===============================
images, labels = next(train_generator)
print("\nBatch Shape:", images.shape)
print("Pixel Range:", images.min(), "to", images.max())
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)
```

```python
base_model.trainable = False
```

```python
#Build model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax')
])
```

```python
#Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#train model
history = model.fit(
    train_generator,  
    validation_data=val_generator,
    epochs=50
)
```

```python
model.save("plant_disease.keras")
```
