```python
import pandas as pd
import random

# =========================
# DISEASES + SYMPTOMS
# =========================
disease_dict = {
    "Tomato Early Blight": ["brown spots", "circular lesions", "dry patches"],
    "Tomato Late Blight": ["water soaked lesions", "black rot", "leaf decay"],
    "Tomato Yellow Leaf Curl Virus": ["yellowing", "leaf curl", "stunted growth"],
    "Potato Early Blight": ["brown concentric rings", "dry spots"],
    "Potato Late Blight": ["dark lesions", "rotting leaves"],
    "Apple Scab": ["olive green spots", "velvety lesions"],
    "Apple Powdery Mildew": ["white powder", "fungal growth"],
    "Grape Black Rot": ["black spots", "leaf lesions"],
    "Grape Powdery Mildew": ["white coating", "powdery growth"],
    "Corn Rust": ["rust pustules", "reddish spots"],
    "Rice Brown Spot": ["brown streaks", "leaf lesions"],
    "Banana Sigatoka": ["black streaks", "yellow spots"],
    "Mango Anthracnose": ["black lesions", "fruit rot"],
    "Cucumber Downy Mildew": ["yellow patches", "angular spots"],
    "Strawberry Leaf Scorch": ["red spots", "leaf burn"],
    "Soybean Rust": ["rust spots", "leaf discoloration"],
    "Coffee Leaf Rust": ["orange spots", "leaf drop"],
    "Chili Leaf Curl Virus": ["leaf curling", "yellowing"],
    "Cotton Leaf Curl Virus": ["leaf curl", "distortion"]
}

# =========================
# TEMPLATES
# =========================
templates = [
    "plant leaves show {}",
    "leaves have {}",
    "symptoms include {} on leaves",
    "plant affected with {}",
    "visible {} on plant leaves",
    "leaf showing {} symptoms",
    "clear signs of {}",
]

# =========================
# GENERATE DATA
# =========================
data = []

samples_per_class = 50   # 50 × ~20 diseases = 1000 rows

for disease, symptoms in disease_dict.items():
    for _ in range(samples_per_class):
        symptom = random.choice(symptoms)
        template = random.choice(templates)
        text = template.format(symptom)
        data.append([text, disease])

# =========================
# SAVE CSV
# =========================
df = pd.DataFrame(data, columns=["text", "label"])
df.to_csv("plant_data.csv", index=False)

print("✅ Dataset created!")
print("Total samples:", len(df))
print(df.head())
```
